# Ausführen des Skripts patch-mmos.js
Dieses Notebook ermöglicht das Ausführen von Shell-Befehlen und externen Skripten im Projektverzeichnis.

## 1. Importieren der benötigten Bibliotheken
Importiere Standardbibliotheken wie `os` und `subprocess` für die Ausführung von Befehlen.

```python
import os
import subprocess
from pathlib import Path

project_root = Path('/workspaces/MecklenburgMarketingOS')
print('Projektverzeichnis:', project_root)
```

## 2. Shell-Befehle im Notebook ausführen
Zeige, wie man mit `!` oder `%%bash` Shell-Befehle direkt im Notebook ausführt.

```python
%%bash
cd /workspaces/MecklenburgMarketingOS
pwd
ls -1 | head -10
```

## 3. Python-Code ausführen und Ergebnisse anzeigen
Führe Python-Code aus und zeige die Ergebnisse im Notebook an.

```python
print('Python-Ausgabe im Notebook funktioniert.')
print('Vorhandene Datei patch-mmos.js:', (project_root / 'patch-mmos.js').exists())
```

## 4. Externe Skripte ausführen
Nutze `subprocess`, um externe Python- oder Shell-Skripte auszuführen und deren Ausgaben zu erfassen.

```python
result = subprocess.run(['bash', 'patch-mmos.js'], cwd=project_root, capture_output=True, text=True)
print('Exit-Code:', result.returncode)
print('STDOUT:
', result.stdout)
print('STDERR:
', result.stderr)
```

## 5. Ausgabe im Output-Pane anzeigen
Demonstriere, wie Ausgaben im Output-Pane von Visual Studio Code angezeigt werden können.

```python
print('Die Ausgabe dieses Codes wird direkt im Notebook-Output angezeigt.')
```

## 6. Fehlerbehandlung beim Ausführen von Code
Implementiere Fehlerbehandlung, um Ausnahmen und Fehlerausgaben beim Ausführen von Code abzufangen.

```python
try:
    result = subprocess.run(['bash', 'patch-mmos.js'], cwd=project_root, capture_output=True, text=True, check=True)
    print('Skript erfolgreich ausgeführt, Ausgabe:')
    print(result.stdout)
except subprocess.CalledProcessError as exc:
    print('Fehler beim Ausführen des Skripts:')
    print('Exit-Code:', exc.returncode)
    print('STDOUT:
', exc.stdout)
    print('STDERR:
', exc.stderr)
```
